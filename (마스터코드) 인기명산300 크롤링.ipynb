```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import subprocess
import shutil
import time
import pandas as pd
from selenium.webdriver import ActionChains
import chromedriver_autoinstaller
import pyperclip
import requests
import lxml
from lxml.html import fromstring
from fake_useragent import UserAgent
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
```

```python
url_list = []
```

```python
url = 'http://www.koreasanha.net/menu/top-100-ranking.htm'

res = requests.get(url) 
parser = fromstring(res.text)

mountain_list = parser.xpath('//*[@id="menu_c"]/div[4]/table/tbody/tr')
print(len(mountain_list))
for i in mountain_list:
    plus_url = i.xpath('.//td[2]/a')[0].get('href')
    url_list.append('http://www.koreasanha.net'+ plus_url)
```

```python
url_2 = 'http://www.koreasanha.net/menu/top-200-ranking.htm'

res = requests.get(url_2) 
parser = fromstring(res.text)

mountain_list = parser.xpath('//*[@id="menu_c"]/div[4]/table/tbody/tr')
print(len(mountain_list))
for i in mountain_list:
    plus_url = i.xpath('.//td[2]/a')[0].get('href')
    url_list.append('http://www.koreasanha.net'+ plus_url)
```

```python
url_3 = 'http://www.koreasanha.net/menu/top-300-ranking.htm'

res = requests.get(url_3) 
parser = fromstring(res.text)

mountain_list = parser.xpath('//*[@id="menu_c"]/div[4]/table/tbody/tr')
print(len(mountain_list))
for i in mountain_list:
    plus_url = i.xpath('.//td[2]/a')[0].get('href')
    url_list.append('http://www.koreasanha.net'+ plus_url)
```

```python
url_list
```

```python
title = []
height = []
location = []
all_contents = []

cnt = 1
for url in url_list:
    try:
        r = requests.get(url) 
        soup = BeautifulSoup(r.content, 'html.parser')

        try:
            title.append(soup.find('h2').text)
        except:
            title.append('')
        
        try:
            height.append(soup.find('p').text.replace('\r','').replace('\n','').split('위치')[0].replace(';','').replace(':','').replace('높이','').lstrip())
        except:
            height.append('')
        try:
            location.append(soup.find('p').text.replace('\r','').replace('\n','').split('위치')[1].replace(';','').replace(':','').lstrip())
        except:
            location.append('')
        try:
            a = ''
            for i in soup.findAll('p'):
                a += i.text
            all_contents.append(a.split('특징, 볼거리')[1].replace('특징, 볼거리', '').replace('산림청 선정 100대 명산', '').split('수정, 보완,')[0].replace('\r','').replace('\n',''))
            print(cnt)
        except:
            all_contents.append('')
    except:
        pass
    cnt += 1
```

```python
final_df = pd.DataFrame(columns = ['산 이름', '높이', '지역', '설명'])
```

```python
final_df['산 이름'] = title
final_df['높이'] = height
final_df['지역'] = location
final_df['설명'] = all_contents
```
