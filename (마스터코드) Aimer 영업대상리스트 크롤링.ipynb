```python
import sys
import re
```

```python
def isHangul(text):
    #Check the Python Version
    pyVer3 =  sys.version_info >= (3, 0)

    if pyVer3 : # for Ver 3 or later
        encText = text
    else: # for Ver 2.x
        if type(text) is not unicode:
            encText = text.decode('utf-8')
        else:
            encText = text

    hanCount = len(re.findall(u'[\u3130-\u318F\uAC00-\uD7A3]+', encText))
    return hanCount > 0
```

```python
real_final = df[~df.인터넷도메인.apply(lambda x: isHangul(x))]
```

```python
real_final = real_final.reset_index()
```

```python
real_final['url'] = real_final.인터넷도메인.apply(lambda x: x.lower())
```

```python
real_final_2 = real_final[real_final.url_2.apply(lambda x: x in sl)].reset_index(drop=True)
```

```python
title_list = []
number_list = []
phone_list = []
address_list = []
email_list = []
final_url = []
sangho_list = []
#################

face_list = []
google_list = []
godomol_list = []
cafe24_list = []

#################
error_sangho_list = []
sangho_error_url = []


start = time.time()

for i in range(14000, len(df)):
    
    url = real_final_2['url_2'][i]
    sangho = real_final_2['상호'][i]
    number = real_final_2['사업자등록번호'][i]
    phone = real_final_2['전화번호'][i]
    address = real_final_2['사업장소재지(도로명)'][i]
    email = real_final_2['전자우편'][i]
    
    try:    
        html = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, timeout = 5)
        tree = fromstring(html.content)
        try:
            title = tree.findtext('.//title')
        except:
            title = ''
        
        if title == None:
            html = requests.get(url+'/site/', headers={"User-Agent": "Mozilla/5.0"})
            time.sleep(0.5)
            tree = fromstring(html.content)
            try:
                title = tree.findtext('.//title')
            except:
                title = ''
            face = 'connect.facebook.net' in html.text
            google = 'googletagmanager.com' in html.text
            godomol = '고도몰' in html.text
            cafe24 = 'cafe24' in html.text
        else:
            face = 'connect.facebook.net' in html.text
            google = 'googletagmanager.com' in html.text
            godomol = '고도몰' in html.text
            cafe24 = 'cafe24' in html.text
            
        final_url.append(url)
        sangho_list.append(sangho)
        title_list.append(title)
        face_list.append(face)
        google_list.append(google)
        number_list.append(number)
        phone_list.append(phone)
        address_list.append(address)
        email_list.append(email)
        godomol_list.append(godomol)
        cafe24_list.append(cafe24)
        
        print(i, '번째 :', sangho, ' ---> 성공')
        
        
    except: ### 로드가 오래걸리는 페이지
        try:
            html = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, timeout = 5)
            tree = fromstring(html.content)
            try:
                title = tree.findtext('.//title')
            except:
                title = ''
                
            if title == None:
                html = requests.get(url+'/site/', headers={"User-Agent": "Mozilla/5.0"})
                time.sleep(0.5)
                tree = fromstring(html.content)
                try:
                    title = tree.findtext('.//title')
                except:
                    title = ''
                
                face = 'connect.facebook.net' in html.text
                google = 'googletagmanager.com' in html.text
                godomol = '고도몰' in html.text
                cafe24 = 'cafe24' in html.text
            else:
                face = 'connect.facebook.net' in html.text
                google = 'googletagmanager.com' in html.text
                godomol = '고도몰' in html.text
                cafe24 = 'cafe24' in html.text
        
            final_url.append(url)
            sangho_list.append(sangho)
            title_list.append(title)
            face_list.append(face)
            google_list.append(google)
            number_list.append(number)
            phone_list.append(phone)
            address_list.append(address)
            email_list.append(email)
            godomol_list.append(godomol)
            cafe24_list.append(cafe24)
        
            print(i, '번째 :', sangho, ' ---> 성공')
                
        except:
            error_sangho_list.append(sangho)
            sangho_error_url.append(url)

            print(i, '번째 :', sangho, ' ---> 실패')
            
sam_df = pd.DataFrame({'상호': sangho_list,'title': title_list, '사업자번호' : number_list
                               ,'연락처': phone_list, '주소' : address_list, '이메일 정보' : email_list,
                               'google': google_list, 'facebook': face_list, '고도몰' : godomol_list,
                               'cafe24' : cafe24_list, 'url': final_url})
        
sam_df.to_csv('2차_sam_df_{}.csv'.format(i), encoding = 'utf-8-sig', index=False)

error_df = pd.DataFrame({'error_place': error_sangho_list, 'error_url' : sangho_error_url})
error_df.to_csv('2차_error_df_{}.csv'.format(i), encoding = 'utf-8-sig', index=False)
```
